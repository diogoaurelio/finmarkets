```python
import requests
from bs4 import BeautifulSoup
import re
import string
```

```python
web = 'https://tradingeconomics.com/indicators'
te = requests.get(web).text
soup = BeautifulSoup(te,'lxml')
items = soup.find_all(class_ = 'list-unstyled')
```

```python
ignoreList = ['Markets','News']
list_category_indic = []
for item in items:
    listElems = item.find_all('li')
    header = listElems[0].text.strip()
    otherElems = [(header, x.text.strip()) for x in listElems[1:]]
    #print(header)
    #print(otherElems)
    list_category_indic.append(otherElems) if header not in ignoreList \
        else print('ignoring {}'.format(header))
```
